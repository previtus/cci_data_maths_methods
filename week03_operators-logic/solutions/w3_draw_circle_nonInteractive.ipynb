```
from matplotlib import pyplot as plt
import numpy as np
import math

fig, ax = plt.subplots()
plt.ylim(-10,10)
plt.xlim(-10,10)

def draw_point(x,y):
    plt.scatter(x, y, s=0.1, c="black")

def draw_circle(x,y,r):
    for t in range(0, 360):
        t_radians = float(t) * math.pi / 180.0
        x_point = r * math.cos(t_radians) + x
        y_point = r * math.sin(t_radians) + y
        draw_point(x_point,y_point)

# This is Colab demo so we can't play with the interactive aspect of this ...
draw_circle(x=1.0, y=0.0,r=1.0)

# but we can manually add more of these:
draw_circle(x=3.0, y=0.0,r=1.0)
draw_circle(x=0.0, y=1.0,r=1.0)
draw_circle(x=2.0, y=1.0,r=1.0)
draw_circle(x=4.0, y=1.0,r=1.0)

plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
