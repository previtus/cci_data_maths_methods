# Random functions in Python

Getting one random number:

```
from numpy.random import normal
from numpy.random import uniform

mean = 0
std = 1
sample_normal = normal(loc=mean, scale=std)

a = 0
b = 1
sample_uniform = uniform(a,b)

print(sample_normal)
print(sample_uniform)
```

Visualization of generated numbers on a plot:

```
mean = 0
std = 1
samples_normal = normal(size=5000, loc=mean, scale=std)

from matplotlib import pyplot as plt
plt.hist(samples_normal, bins=50)
plt.show()
```

```
a = 0
b = 1
samples_uniform = uniform(a,b, size=5000)

from matplotlib import pyplot as plt
plt.hist(samples_uniform, bins=50)
plt.show()
```

# Possible solutions

```
size = [500, 500, 3]
empty_image = np.zeros(size)

dots = 99999
for i in range(dots):
    random_integer_1 = int(uniform(0,500))
    random_integer_2 = int(uniform(0,500))
    empty_image[random_integer_1,random_integer_2,:] = 255

IPython.display.display(PIL.Image.fromarray(np.uint8(empty_image)))
```

```
size = [500, 500, 3]
empty_image = np.zeros(size)
mean = 250
std = 10
std = 100

dots = 99999
for i in range(dots):
    random_integer_1 = int(normal(loc=mean, scale = std))
    random_integer_2 = int(normal(loc=mean, scale = std))

    # just to be save, clip them to be in between 0 to 500!
    random_integer_1 = max(0, min(500 -1, random_integer_1))
    random_integer_2 = max(0, min(500 -1, random_integer_2))

    empty_image[random_integer_1,random_integer_2,:] = 255

IPython.display.display(PIL.Image.fromarray(np.uint8(empty_image)))
```
